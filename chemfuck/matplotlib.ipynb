```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def funck(x, offset=10):
  y = np.sin(x) + np.cos(x**2) - np.sqrt(x) + offset
  return y
```

```python
x_plot = [0,1,2,3,4]
y_plot = [0,1,4,9,16]

plt.figure()
plt.plot(x_plot,y_plot)

plt.show()
```

```python
x_plot = np.linspace(0,2*np.pi, 1000)
y_plot = funck(x_plot, 5)
plt.figure()
plt.plot(x_plot,y_plot)
plt.plot(x_plot,2*np.sin(x_plot)+3)
plt.show()
```

```python
x_plot = np.linspace(0,2*np.pi, 1000)
y_plot = funck(x_plot, 5)
x1_plot = [0,1,2,3,4]
y1_plot = [0,1,4,9,6]

plt.figure(figsize=(6,6))

plt.plot(x_plot,y_plot, 'r-', label="funck")
plt.plot(x_plot,2*np.sin(x_plot)+3, 'b:', label="sin")
plt.plot(x1_plot,y1_plot, 'go', label="line")

plt.xlim(0,2*np.pi)
plt.ylim(0,7)

plt.xlabel("Чиселки снизу")
plt.ylabel("Чиселки слева")
plt.title("Чиселки повсюду")

plt.tick_params(direction="in")

plt.grid(True)

plt.legend()

plt.savefig("plot1.png", dpi=500)

plt.show()
```
