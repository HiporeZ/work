Групировка данных

выполняет что-то для отдельных групп

```python
df.groupby(["type"])[["pH",'density',"quality","residual sugar"]].agg(
    [np.mean, np.std, np.min, np.max]
)
```

```python
df[df['residual sugar'].isnull()]
```

```python
print(df.isnull().sum())
print(df.isnull().sum().sum())
```
